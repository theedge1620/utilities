```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
  
# Read Images
img = mpimg.imread('./img_processing/botw_guardian.jpg')
  
# Output Images
plt.imshow(img)
```

```python
img.shape
```

```python
img[0]
```

```python
M = img.shape[0]//3
N = img.shape[1]//3
tiles = [img[x:x+M,y:y+N] for x in range(0,img.shape[0],M) for y in range(0,img.shape[1],N)]
```

```python
for ii in range(0,len(tiles)):
    fname = 'botw_guardian_large_00'+str(ii+1)+'.png'
    mpimg.imsave(fname, tiles[ii])
```
